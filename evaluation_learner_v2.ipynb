The Soundfile dataset was splitet in to 5 Seconds, 3 Seconds and 1 Second fragments and convertet to spectograms. For the 5 second fragments some single files got removed from dataset because they are shorter then 5 seconds. (these files are also removed from 1 second dataset so the input soundfiles of the 3 second dataset are a bit different but this will be neglected because of the small quantity of removed files).

The Datasets got split into learn (70%), validation(20%) and test(10%) datasets by random selection. 

For all learners model resnet 34 with error_rate metrics was used (learn = cnn_learner(data, models.resnet34, metrics=error_rate))
